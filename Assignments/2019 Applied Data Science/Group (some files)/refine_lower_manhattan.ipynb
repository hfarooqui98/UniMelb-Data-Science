```python
import pandas as pd
import numpy as np
from collections import Counter
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
data = pd.read_feather('data/all.feather')
```

```python
loc_groupby = data.groupby(['pickupX', 'pickupY']).size()
```

```python
loc_df = loc_groupby.rename('counts').reset_index()
```

## Try plotting with different normalisation values for number of counts.

```python
plt.figure(figsize=(8, 10))
plt.gca().invert_yaxis()
sns.scatterplot(
    x='pickupX',
    y='pickupY',
    data=loc_df,
    s=20,
    hue='counts',
    hue_norm=(0, 1000000/1),
)
```

```python
plt.figure(figsize=(8, 10))
plt.gca().invert_yaxis()
sns.scatterplot(
    x='pickupX',
    y='pickupY',
    data=loc_df,
    s=20,
    hue='counts',
    hue_norm=(0, 1000000/2),
)
```

## Get Lower Manhattan Cell Manually

With eyeballing, it is fairly easy to see which part are getting most of trips. Exclude these airport cell, we assume the other ones are lower Manhattan.

The threshold used here is 400,000 counts of trips in the cell amoung the entire time of training dataset.

```python
manhattan_count = loc_df[(loc_df.counts > 400000) & (loc_df.pickupX < 30)]
```

```python
plt.figure(figsize=(8, 10))
plt.gca().set_xlim((loc_df.pickupX.min(), loc_df.pickupX.max()))
plt.gca().set_ylim((loc_df.pickupY.min(), loc_df.pickupY.max()))
plt.gca().invert_yaxis()
sns.scatterplot(
    x='pickupX',
    y='pickupY',
    data=manhattan_count,
    s=20,
    hue='counts',
)
```

```python
manhattan_cells = manhattan_count[['pickupX', 'pickupY']].values
```
